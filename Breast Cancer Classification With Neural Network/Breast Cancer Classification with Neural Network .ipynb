```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Data collection and Processing 
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the target column 

data_frame['label'] = breast_cancer_dataset.target
```

```python
# Note, 1 represent the people that has people that has breast cancer and 0 represent people with no breast cancer 
```

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns ='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state= 2)
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train.shape, X_test.shape, X.shape)
```

```python
# Building the Neural Network
```

```python
# importing tensorflow and keras
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from keras import Input, Sequential
from keras.layers import Flatten, Dense
```

```python
model = Sequential([
    Input(shape=(30,)),  # This is the proper way to define the input shape
    Flatten(),  # Flatten layer (input layer)
    Dense(20, activation='relu'),  # Hidden layer
    Dense(2, activation='sigmoid')  # Output layer
])
```

```python
# compiling the Neural Network 

model.compile(optimizer = 'adam',
              loss ='sparse_categorical_crossentropy',
              metrics =['accuracy']
             )
```

```python
# training the Neural Network 

history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs =10)
```

```python
# visualizing the accuracy and loss
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model accuracy')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
# Accuracy of the model on test data
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
